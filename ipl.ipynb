```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pt

df = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
```

```python
features = df[["bowler","extra_runs"]]
pt.plot(features["bowler"],features["extra_runs"])
pt.show()
df.head()
```

```python
#features = df[["bowler","over_wicket"]]

df["over_wicket"] = df.loc[df["is_wicket"] !== 0]
features = df[["bowler","over_wicket"]]
pt.plot(features["bowler"],features["over_wicket"])
pt.show()
df.head(30)
```

```python
df1 = pd.read_csv("IPL Matches 2008-2020.csv")
df1.sort_values(by=["result_margin"], inplace=True)
features = df1[["result_margin","team1"]]
pt.plot(features["result_margin"],features["team1"])
pt.show()
```
